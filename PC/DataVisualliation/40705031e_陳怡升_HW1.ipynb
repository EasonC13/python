```python
#Calculate each team’s average ERA from 2011 to 2015, 
#and list the result as a table, where each column is a year, each row is a team.
df = pd.read_csv("./Data/pitching.csv")

dfs = []
for i in range(2011,2016):
    out = df[df.yearID == i].groupby("teamID").ERA.mean()
    out = pd.DataFrame(out).rename(columns = {"ERA":i})
    dfs.append(out)

pd.concat(dfs, axis = 1, sort = True)
```

```python
#Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015,
#and list the result as a table, where each column is a league and each row is a year.

df = pd.read_csv("./Data/pitching.csv")

target = ["AL", "NL"]
years = [2011, 2012, 2013, 2014, 2015]
outdf = pd.DataFrame(columns=years, index=target)

for igId in target:
    df2 = df[df.lgID == igId]
    for year in years:
        out = df2[df2.yearID == year].ERA.mean()
        outdf[year][igId] = out
outdf
```
