```python
import os
import glob
import numpy as np
from keras.preprocessing.image import  img_to_array, load_img
from PIL import Image
```

```python
dict_labels = {"drawings":0, "hentai":1, "neutral":2, "porn": 3, "sexy": 4}
size = (300,300)
nbofdata=500
```

```python
test = 0
test2 = 0
check = 1
nbofdata = 500
less_nbofdata = 0
i_save = 0
while(1):
    for folders in glob.glob("./train/*"):
        print()
        print(folders)
        images=[]
        labels_hot=[]
        labels=[]
        nbofdata_i=1
        for filename in os.listdir(folders):
            test+=1
            if nbofdata_i <= nbofdata and nbofdata_i >less_nbofdata:
                test2+=1
            #print(filename)
                label = os.path.basename(folders)
                className = np.asarray(label)
                img=load_img(os.path.join(folders,filename))
                img=img.resize(size,Image.BILINEAR)
                if img is not None:
                    if label is not None:
                        labels.append(className)
                        labels_hot.append(dict_labels[label])
                    x=img_to_array(img)
                    images.append(x)
                i_save = nbofdata_i
            nbofdata_i+=1
                #print("Out of range!!")
            print("test = %d, test2 = %d, i is %d, nbofdata is %d, less_nbofdata is %d"
                  %(test,test2, i_save, nbofdata,less_nbofdata) , end='\r')
            ##save
        images=np.array(images)
        print("test = %d, test2 = %d, i is %d, nbofdata is %d, less_nbofdata is %d"
                  %(test,test2, i_save, nbofdata,less_nbofdata))
        labels_hot=np.array(labels_hot)
        print("images.shape={}, labels_hot.shape=={}".format(images.shape, labels_hot.shape))    
        imagesavepath='NSFW_Dataset/train/'+str(less_nbofdata+1)+"_"+str(nbofdata)+"/"+label+"/"
        if not os.path.exists(imagesavepath):
            os.makedirs(imagesavepath)
        np.save(imagesavepath+'{}_images.npy'.format(label)+str(less_nbofdata)+"_"+str(nbofdata),images)    
        np.save(imagesavepath+'{}_label.npy'.format(label)+str(less_nbofdata)+"_"+str(nbofdata),labels)    
        np.save(imagesavepath+'{}_labels_hot.npy'.format(label)+str(less_nbofdata)+"_"+str(nbofdata),labels_hot)
        print('{} files has been saved.'.format(label)+ "range in: " + str(less_nbofdata)+"_"+str(nbofdata))
    if check ==1:
        total = test
        check =0
        print("total is %d"%(total))
    if test2==total:
        break
    ##
    '''
    images=np.array(images)    
    labels_hot=np.array(labels_hot)
    print("images.shape={}, labels_hot.shape=={}".format(images.shape, labels_hot.shape))    
    imagesavepath='NSFW_Dataset/test/'
    if not os.path.exists(imagesavepath):
        os.makedirs(imagesavepath)
    np.save(imagesavepath+'{}_images.npy'.format(label)+nbofdata+"_"+less_nbofdata,images)    
    np.save(imagesavepath+'{}_label.npy'.format(label)+nbofdata+"_"+less_nbofdata,labels)    
    np.save(imagesavepath+'{}_labels_hot.npy'.format(label)+nbofdata+"_"+less_nbofdata,labels_hot)
    print('{} files has been saved.'.format(label)+ "range in: " + nbofdata+"_"+less_nbofdata)
    '''
    if nbofdata_i >= nbofdata:
        #models.fit
        images=[]
        labels_hot=[]
        labels=[]
        nbofdata+=500
        less_nbofdata+=500
        print("NBO更新囉")
```
